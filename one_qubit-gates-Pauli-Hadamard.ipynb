```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from qiskit import *
from math import pi
import numpy as np
from qiskit.visualization import *
import matplotlib.pyplot as plt
```

## Pauli X-gate
The Pauli-X gate is a single-qubit rotation through \piÏ€ radians around the x-axis.\
$$\left|\psi\right\rangle = \alpha\left|0\right\rangle + \beta \left|1\right\rangle$$\
$$
X   =  
\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix}
$$


$$X \left|0\right\rangle = $$
 \
$$\alpha = 1 , \beta = 0 \ \ \ to \ \ \ \alpha = 0 , \beta = 1$$\
and \
$$X \left|1\right\rangle = $$
 \
$$\alpha = 0 , \beta = 1 \ \ \ to \ \ \ \alpha = 1 , \beta = 0$$

### Draw cricuit

```python
qreg_qx = QuantumRegister(1, 'q')
creg_cx = ClassicalRegister(1, 'c')
qc_x = QuantumCircuit(qreg_qx, creg_cx)

qc_x.x(qreg_qx[0])
#qc_x.x(qreg_qx[0])
#qc_x.h(qreg_qx[0])
qc_x.measure(qreg_qx[0], creg_cx[0])
qc_x.draw(output='mpl')
```

### Simulate the circuit and plot statevector in Bloch sphere

```python
backend = Aer.get_backend('statevector_simulator')
out_x = execute(qc_x,backend).result().get_statevector()
print(out_x)
plot_bloch_multivector(out_x)
```

### Display phase in Bloch sphere

```python
plot_state_qsphere(out_x)
```

## Hadamard gate,  H-gate
$$
H   =  
\frac{1}{\sqrt{2}}\begin{bmatrix}
1 & 1\\
1 & -1
\end{bmatrix}
$$

$|0 \rangle =  \frac{1}{\sqrt{2}} |0\rangle  + \frac{1}{\sqrt{2}} |1 \rangle$ \
and\
$|1 \rangle =  \frac{1}{\sqrt{2}} |0\rangle  - \frac{1}{\sqrt{2}} |1 \rangle$\

$$\left|\psi\right\rangle = \alpha\left|0\right\rangle + \beta \left|1\right\rangle$$

$$H\left|\psi\right\rangle = \alpha(\frac{1}{\sqrt{2}} |0\rangle  + \frac{1}{\sqrt{2}} |1 \rangle) + \beta(\frac{1}{\sqrt{2}} |0\rangle  - \frac{1}{\sqrt{2}} |1 \rangle)$$ 

$$H\left|\psi\right\rangle = \frac{1}{\sqrt{2}}[(\alpha + \beta )|0\rangle  +  (\alpha - \beta)|1 \rangle)]$$ \
Double operation by H,
$$H{H\left|\psi\right\rangle} = H({\frac{1}{\sqrt{2}}[(\alpha + \beta )|0\rangle  +  (\alpha - \beta)|1 \rangle)]})$$ \
$$ = {\frac{1}{\sqrt{2}}[(\alpha + \beta )(\frac{1}{\sqrt{2}} |0\rangle  + \frac{1}{\sqrt{2}} |1 \rangle)  +  (\alpha - \beta)(\frac{1}{\sqrt{2}} |0\rangle  - \frac{1}{\sqrt{2}} |1 \rangle))]} = \left|\psi\right\rangle$$
