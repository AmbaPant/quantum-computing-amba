### Some ideas to start research on quantum computing using IBM online quantum computers

1. There are many **experimental data** in different materials available in literatures and **Monte Carlo simulation/DFT calculations** using classical computers/super computers. Prepare programs/codes on IBM QC and **compare results**.
- H dynamics in materials
- Electron/ion behavior in battery materials
- Ion movements in lithium battery 
- Nanoscience properties
- A very simple system containing H and/or H2O


2. **Spin dynamics** on/across sub-surface/boundary conditions
- Silicon / semiconductors 
- Superconductors


3. Search of **new functional materials** 
- Simulation with manipulation of materials and find new properties
- Materials for quantum computers 
- Materials for green energy sources

4. Entanglement
- Quantum enganglement in materials
- Biological entanglement
- etc.
